```python
import os
import shutil
# tässä siirretään luokitellut kuvat pois koulutusta varten.

folder_a = 'data/kuvat' # kaikki 10000~ kuvaa
folder_b = 'train_data/images/train' # train kuvat Koulutuskuvien tiedostopolku
folder_c = 'train_data/images/val' # val kuvat Validointikuvien tiedostopolku

file_list_path1 = 'data/train.txt'
file_list_path2 = 'data/val.txt'

def move_files(From, To, file_names):
    """
    Siirtää tiedostot listan mukaisesti kansiosta toiseen.
    
    :param From: Lähdekansio
    :param To: Kohdekansio
    :param file_names: Lista tiedostonimistä
    """

    for f in file_names:
        source_path = os.path.join(From, f.strip())
        destination_path = os.path.join(To, f.strip())

        if os.path.exists(source_path):
            shutil.move(source_path, destination_path)
            print(f"moved {f} to {To}")
        else:
            print(f"file {f} not found in {From}")
```

```python
# Siirretään koulutuskuvat
with open(file_list_path1, 'r') as file1:
    file_names1 = file1.readlines()
move_files(folder_a, folder_b, file_names1)

# Siirretään validointikuvat
with open(file_list_path2, 'r') as file2:
    file_names2 = file2.readlines()
move_files(folder_a, folder_c, file_names2)
```
